### Categorical Plots


####  Creating a categorical variable that splits the "Price per Unit" column into categories.

```python
sns.histplot(df['2nd_dose_per_hundred'], bins = 20, kde = True)
```

```python
df.loc[df['2nd_dose_per_hundred'] < 20, '2nd_dose_category'] = 'Low_amount'
```

```python
df.loc[(df['2nd_dose_per_hundred'] >=20) & (df['2nd_dose_per_hundred'] <50), '2nd_dose_category'] = 'Middle_amount'
```

```python
df.loc[df['2nd_dose_per_hundred'] >= 50, '2nd_dose_category'] = 'High_amount'
```

```python
df['2nd_dose_category'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the 2nd_dose_per_hundred categories created above

sns.set(style="ticks", rc = {'figure.figsize':(9,6)})
g = sns.stripplot(x= '1st_dose_per_hundred', y='2nd_dose_per_hundred', hue='2nd_dose_category', data=df.sample(400))
```

2nd_dose variables shows that fully vaccinated people. The countries with people who vaccinated 1st_dose high amount, they are more closely to fully vaccinated because they have more people with 2nd_dose.
